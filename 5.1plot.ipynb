```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned CSV file into a DataFrame
csv_file = "gyroscope_data_cleaned.csv"
df = pd.read_csv(csv_file)

# Display the first few rows of the DataFrame
print(df.head())

# Convert the 'Timestamp' column to datetime for better plotting
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Plot each variable separately
plt.figure(figsize=(12, 8))

# Plot for X-axis
plt.subplot(3, 1, 1)
plt.plot(df['Timestamp'], df['X-axis (xx)'], color='r', label='X-axis (xx)')
plt.xlabel('Timestamp')
plt.ylabel('X-axis (xx)')
plt.title('X-axis (xx) Over Time')
plt.legend()

# Plot for Y-axis
plt.subplot(3, 1, 2)
plt.plot(df['Timestamp'], df['Y-axis (yy)'], color='g', label='Y-axis (yy)')
plt.xlabel('Timestamp')
plt.ylabel('Y-axis (yy)')
plt.title('Y-axis (yy) Over Time')
plt.legend()

# Plot for Z-axis
plt.subplot(3, 1, 3)
plt.plot(df['Timestamp'], df['Z-axis (zz)'], color='b', label='Z-axis (zz)')
plt.xlabel('Timestamp')
plt.ylabel('Z-axis (zz)')
plt.title('Z-axis (zz) Over Time')
plt.legend()

# Adjust layout
plt.tight_layout()
plt.show()

# Plot all three variables in a single graph
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['X-axis (xx)'], color='r', label='X-axis (xx)')
plt.plot(df['Timestamp'], df['Y-axis (yy)'], color='g', label='Y-axis (yy)')
plt.plot(df['Timestamp'], df['Z-axis (zz)'], color='b', label='Z-axis (zz)')
plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.title('Gyroscope Data Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
